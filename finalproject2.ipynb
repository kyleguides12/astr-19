```python
# Import
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

# paths
f105w_file = 'hlsp_hudf12_hst_wfc3ir_udfmain_f105w_v1.0_drz (2).fits'
f125w_file = 'hlsp_hudf12_hst_wfc3ir_udfmain_f125w_v1.0_drz.fits'
f160w_file = 'hlsp_hudf12_hst_wfc3ir_udfmain_f160w_v1.0_drz.fits'

# Label
image_f105w = fits.getdata(f105w_file)
image_f125w = fits.getdata(f125w_file)
image_f160w = fits.getdata(f160w_file)

# Normalize the images 
image_f105w_normalized = (image_f105w - np.min(image_f105w)) / (np.max(image_f105w) - np.min(image_f105w))
image_f125w_normalized = (image_f125w - np.min(image_f125w)) / (np.max(image_f125w) - np.min(image_f125w))
image_f160w_normalized = (image_f160w - np.min(image_f160w)) / (np.max(image_f160w) - np.min(image_f160w))

# Stack to form RGB Image
rgb_image = np.dstack((image_f160w_normalized, image_f125w_normalized, image_f105w_normalized))

# individual images
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

axes[0].imshow(image_f160w_normalized, cmap='Reds')  # F160W in red
axes[0].set_title('F160W (Red)')
axes[0].axis('off')  

axes[1].imshow(image_f125w_normalized, cmap='Greens')  # F125W in green
axes[1].set_title('F125W (Green)')
axes[1].axis('off')

axes[2].imshow(image_f105w_normalized, cmap='Blues')  # F105W in blue
axes[2].set_title('F105W (Blue)')
axes[2].axis('off')

# individual images
plt.show()

# Plot combined image
plt.imshow(rgb_image)
plt.title('Hubble Ultra Deep Field False-Color Image (F160W, F125W, F105W)')
plt.axis('off')  # Hide axis labels

# PNG File
plt.savefig('hubble_udf_false_color_image.png', dpi=300)


plt.show()
```
